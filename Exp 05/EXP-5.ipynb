```python
import pandas as pd
import matplotlib.pyplot as plt
#load the data set 
data = pd.read_csv(r"facebook.csv")
#return the first few rows of the file
print(data.head()) 
```

```python
#mean
mean = data['age'].mean()
print(f"Mean actual age: {mean}")
```

```python
#meadian
median = data['age'].median()
print(f"Median actual age: {median}")
```

```python
#mode
mode = data['age'].mode()[0]
print(f"Mode actual age: {mode}")
```

```python
#Variance
variance =data['age'].var()
print(f"Variance actual age: {variance}")
```

```python
# Standard deviation (Formula: square root of the variance)
std_dev = data['age'].std()
print(f"Standard Deviation actual age: {std_dev}")
```

```python
#histogram is useful for visualizing the distribution of age data
plt.figure(figsize=(10, 6))
plt.hist(data['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
