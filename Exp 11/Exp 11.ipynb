```python
# Import necessary libraries
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
# Define the features and labels
features = np.array([[150, 7], [160, 7.5], [170, 8], [140, 6.5], [130, 6], [180, 8.5], [198, 9], [120, 5.5]])
labels = np.array([0, 0, 0, 0, 0, 1, 1, 1])

# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Create the model using 3 nearest neighbors
knn = KNeighborsClassifier(n_neighbors=3)

# Train the model
knn.fit(X_train, y_train)

# Make predictions
y_pred = knn.predict(X_test)
```

```python
# Print the predicted and actual labels
print("Predicted Labels:", y_pred)
print("Actual Labels:", y_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Test on a new fruit (apple with 155g weight and 7cm diameter)
new_fruit = [[155, 7]]
prediction = knn.predict(new_fruit)

# Print the new fruit prediction (0: Apple, 1: Orange)
print("New Fruit Prediction (0 = Apple, 1 = Orange):", prediction[0])
```

```python
# Visualization of the results
# Separate apples and oranges for visualization
apples = features[labels == 0]
oranges = features[labels == 1]

# Plot data points
plt.scatter(apples[:, 0], apples[:, 1], color='red', label="Apple", marker="o")
plt.scatter(oranges[:, 0], oranges[:, 1], color='orange', label="Orange", marker="x")

# Plot the new fruit
plt.scatter(new_fruit[0][0], new_fruit[0][1], color="blue", label='New Fruit', marker='*', s=150)

# Labels and Legend
plt.title("KNN Classification: Apple vs Orange")
plt.xlabel("Weight (grams)")
plt.ylabel("Size (diameter in cm)")
plt.legend()
plt.grid(True)
plt.show()
```
