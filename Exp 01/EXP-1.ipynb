```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data=pd.read_csv("amazon.csv")
print(data.head())

# Converting categorical data into numeric data
data['rating_count'] = pd.to_numeric(data['rating_count'], errors='coerce')
```

```python
# Calculating mean, median, and mode on the elements of "rating_count" column
mean = data['rating_count'].mean()
print(f"Mean actual price: {mean}")

median_actual_price = data['rating_count'].median()
print(f"Median actual price: {median_actual_price}")

mode_actual_price = data['rating_count'].mode()
print(f"Mode actual price: {mode_actual_price}")
```

```python
# Histogram of price columnn
plt.figure(figsize=(10,6))
sns.histplot(data['rating_count'], bins=20, kde=True)
plt.title('Rating Count')
plt.show()
```

```python
# Boxplot of the ratings
plt.figure(figsize=(10, 6))
sns.boxplot(x = data['rating'])
plt.title('Ratings Distribution')
plt.show()
```

```python
# Scatterplot of price vs. units sold
plt.figure(figsize = (10, 6))
sns.scatterplot(x='discounted_price', y='actual_price', data=data)
plt.title('Price vs. Units Sold')
plt.show()
```

```python
# Barplot of categories vs. average units sold
plt.figure(figsize=(10, 6))
sns.barplot(x='category', y='discounted_price', data = data)
plt.title('Average Units sold by Category')
plt.xticks(rotation = 90) # Rotate x-axis label for better readability
plt.show()
```
